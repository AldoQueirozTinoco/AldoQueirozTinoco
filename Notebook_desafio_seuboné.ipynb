##Para o restante das tabelas:

Agora que fiz o processo para 1, hora de fazer uma função para abarcar o resto das tabelas (Que já estão formatadas similarmente)

````python
def top3(df,ano):
  # Note que:
  # A função abaixo é apenas uma mistura dos passos de 1 a 4 feitos na tabela de 2020
  # EXP_2020.groupby(['Estado','Produto'])['Valor'].sum() + .reset_index() + .sort_values(['Estado','Valor'],ascending=[True,False])
  maisVendidosPorEstado=df.groupby(['Estado','Produto'])['Valor'].sum().reset_index().sort_values(['Estado','Valor'],ascending=[True,False])
  top3_ano = maisVendidosPorEstado.groupby('Estado').head(3)
  return top3_ano

top3_IMP_2020 = top3(IMP_2020,'2020')
top3_EXP_2021 = top3(EXP_2021,'2021')
top3_IMP_2021 = top3(IMP_2021,'2021')
````

Código completo para tratamento da tabela NCM:

````python
# Descrições que não ajudam a indentificar o produto, utilizadas para substituir estas por descrições melhores adiante
desc_irrelevantes=['Outros', '-Outros', '--Outros','- Outros','-- Outros', 'Outras', '-Outras', '--Outras','- Outras','-- Outras']

# Crio uma nova coluna apenas com os códigos de 2 digitos (Antes de preencher com zeros a esquerda)
ncm_df['Código_2digitos'] = ncm_df['Código'].str[:2]

# Seleciona códigos com exatamente 2 caracteres e mantém apenas código e descrição dessas linhas
# Renomeia Descrição para Descrição_Genérica
ncm_generico = ncm_df[ncm_df['Código'].str.len()==2][['Código','Descrição']].rename(columns={'Descrição': 'Descrição_Genérica'})

# Preenchemos os espaços a esquerda vazios com zero até o numero ter 8 algarismos
# Fazemos isso pois o formato dos códigos das tabelas do GOV e do site da fazenda são diferentes e para podermos dar merge nas 2 tabelas
ncm_df['Código'] = ncm_df['Código'].str.replace('.','').str.zfill(8)

# Juntamos as tabelas ncm com ncm_generico
# Basicamente anexamos a coluna 'Descrição_Genérica' a tabela ncm_df
ncm_df = ncm_df.merge(ncm_generico,left_on='Código_2digitos',right_on='Código',how='left',suffixes=('','_Genérica'))

# Utilizo a fução apply em conjunto com uma função lambda para por um if else na descrição do elemento,
# se o elemento é nulo ou está tido como alguma variação de 'Outro' será substituido pela descrição genérica
ncm_df['Descrição_para_merge'] = ncm_df.apply(lambda row: row['Descrição_Genérica']
    if pd.isna(row['Descrição']) or str(row['Descrição']).strip() in desc_irrelevantes
    else row['Descrição'],
  axis=1
)
````

### Para os dados de 2020:

Agora basta pegar a tabela e:
1. Padronizar os códigos de produto
2. Remover colunas redundantes
3. Fazer o Merge com a tabela NCM
4. Criar de uma coluna auxiliar com os 2 primeiros dígitos do produto
5. Fazer um dictionary para lidar com valores NaN
6. Substituir os valores NaN em Descrição

````python
top3_EXP_2020.head(8)
````

````python
#Pondo zeros a esquerda para fazer o merge
top3_EXP_2020['Produto'] = top3_EXP_2020['Produto'].astype(str).str.replace('.', '').str.zfill(8)

# Remoção da coluna 'Descrição' que será recriada no merge
top3_EXP_2020 = top3_EXP_2020.drop(columns=['Descrição'], errors='ignore')

# Merge da tabela de 2020 com a tabela NCM e renomeando 'Descrição_para_merge' como 'Descrição'
df_EXP_2020_merged = top3_EXP_2020.merge(
    ncm_df[['Código', 'Descrição_para_merge']],
    left_on='Produto',
    right_on='Código',
    how='left'
).rename(columns={'Descrição_para_merge': 'Descrição'})

# Remoção das colunas "Código" e "Nome_Produto" que são redundantes
df_EXP_2020_merged = df_EXP_2020_merged.drop(columns=['Código'])

# Preenchendo os valores NaN da Descrição com base nos dois primeiros dígitos do Produto
# Cria coluna auxiliar com os dois primeiros dígitos do código
df_EXP_2020_merged['Produto_2dig'] = df_EXP_2020_merged['Produto'].str[:2]

# Mapeamento de Código_2digitos -> Descrição_Genérica
mapa_generico = ncm_df.drop_duplicates('Código_2digitos').set_index('Código_2digitos')['Descrição_Genérica'].to_dict()

# Substituição dos NaNs na Descrição com a Descrição Genérica correspondente
df_EXP_2020_merged['Descrição'] = df_EXP_2020_merged.apply(
    lambda row: mapa_generico.get(row['Produto_2dig'], row['Descrição']) if pd.isna(row['Descrição']) else row['Descrição'],
    axis=1
)

# Remove a coluna auxiliar
df_EXP_2020_merged = df_EXP_2020_merged.drop(columns=['Produto_2dig'])
````

### Para o restante das tabelas:

Agora fazemos uma função para aplicar os mesmos métodos para o restante das tabelas.

````python
def aplicar_descricao_ncm_com_generico(df, ncm_df, desc_irrelevantes):
    # Formata a coluna Produto com 8 dígitos, removendo pontos e adicionando zeros à esquerda
    df['Produto'] = df['Produto'].astype(str).str.replace('.', '').str.zfill(8)

    # Preenche valores ausentes da coluna Nome_Produto (se existir)
    if 'Nome_Produto' in df.columns:
        df['Nome_Produto'] = df['Nome_Produto'].fillna('Descrição Genérica')

    # Remove coluna 'Descrição' caso exista
    df = df.drop(columns=['Descrição'], errors='ignore')

    # Merge com descrição limpa da tabela NCM
    df = df.merge(
        ncm_df[['Código', 'Descrição_para_merge']],
        left_on='Produto',
        right_on='Código',
        how='left'
    ).rename(columns={'Descrição_para_merge': 'Descrição'})

    # Remove colunas redundantes
    df = df.drop(columns=['Código'], errors='ignore')
    df = df.drop(columns=['Nome_Produto'], errors='ignore')

    # Preenche NaNs com descrição genérica com base nos dois primeiros dígitos
    df['Produto_2dig'] = df['Produto'].str[:2]
    mapa_generico = (
        ncm_df.drop_duplicates('Código_2digitos')
              .set_index('Código_2digitos')['Descrição_Genérica']
              .to_dict()
    )

    df['Descrição'] = df.apply(
        lambda row: mapa_generico.get(row['Produto_2dig'], row['Descrição']) if pd.isna(row['Descrição']) else row['Descrição'],
        axis=1
    )

    # Remove a coluna auxiliar
    df = df.drop(columns=['Produto_2dig'])

    return df
````

````python
df_EXP_2020_merged = aplicar_descricao_ncm_com_generico(top3_EXP_2020, ncm_df, desc_irrelevantes)
df_IMP_2020_merged = aplicar_descricao_ncm_com_generico(top3_IMP_2020, ncm_df, desc_irrelevantes)
df_EXP_2021_merged = aplicar_descricao_ncm_com_generico(top3_EXP_2021, ncm_df, desc_irrelevantes)
df_IMP_2021_merged = aplicar_descricao_ncm_com_generico(top3_IMP_2021, ncm_df, desc_irrelevantes)
````
